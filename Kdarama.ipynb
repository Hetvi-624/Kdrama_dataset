```python
genre_counts = df["Genre"].dropna().str.split(",").explode().str.strip().value_counts()
genre_df = genre_counts.reset_index()
genre_df.columns = ["Genre", "Count"]
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_df["Count"], y=genre_df["Genre"], palette="viridis")
plt.show()
```

```python
actors_counts = df["Actors"].dropna().str.split(",").explode().str.strip().value_counts()
top_actors = actors_counts.head(30).reset_index()
top_actors.columns = ["Actor", "Count"]
plt.figure(figsize=(12, 6))
sns.barplot(x=top_actors["Count"], y=top_actors["Actor"], palette="coolwarm")
plt.show()
```

#Data Preparation

```python
Q1=df['Number of Episodes'].quantile(0.25)
Q3=df['Number of Episodes'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
df1=df[(df['Number of Episodes']>=lower_bound)&(df['Number of Episodes']<=upper_bound)]
plt.figure(figsize=(10, 5))
sns.boxplot(data=df1, x=df1['Number of Episodes'])
plt.show()
```

```python
Q1=df['Rating'].quantile(0.25)
Q3=df['Rating'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
df2=df[(df['Rating']>=lower_bound)&(df['Rating']<=upper_bound)]
plt.figure(figsize=(10, 5))
sns.boxplot(data=df2, x=df2['Rating'])
plt.show()
```

```python
Q1=df['Year of release'].quantile(0.25)
Q3=df['Year of release'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
df3=df[(df['Year of release']>=lower_bound)&(df['Year of release']<=upper_bound)]
plt.figure(figsize=(10, 5))
sns.boxplot(data=df3, x=df3['Year of release'])
plt.show()
```
